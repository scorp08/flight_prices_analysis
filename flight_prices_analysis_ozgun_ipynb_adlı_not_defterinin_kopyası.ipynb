```python
def iqr_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[column] < lower) | (data[column] > upper)]
outliers_totalfare = iqr_outliers(df_num, "totalFare")
print("Total Fare outlier sayısı:", outliers_totalfare.shape[0])
```

```python
def grab_col_names(dataframe, cat_th=10, car_th=20):
    """

    Veri setindeki kategorik, numerik ve kategorik fakat kardinal değişkenlerin isimlerini verir.
    Not: Kategorik değişkenlerin içerisine numerik görünümlü kategorik değişkenler de dahildir.

    Parameters
    ------
        dataframe: dataframe
                Değişken isimleri alınmak istenilen dataframe
        cat_th: int, optional
                numerik fakat kategorik olan değişkenler için sınıf eşik değeri
        car_th: int, optinal
                kategorik fakat kardinal değişkenler için sınıf eşik değeri

    Returns
    ------
        cat_cols: list
                Kategorik değişken listesi
        num_cols: list
                Numerik değişken listesi
        cat_but_car: list
                Kategorik görünümlü kardinal değişken listesi

    Examples
    ------
        import seaborn as sns
        df = sns.load_dataset("iris")
        print(grab_col_names(df))


    Notes
    ------
        cat_cols + num_cols + cat_but_car = toplam değişken sayısı
        num_but_cat cat_cols'un içerisinde.
        Return olan 3 liste toplamı toplam değişken sayısına eşittir: cat_cols + num_cols + cat_but_car = değişken sayısı

    """

    # cat_cols, cat_but_car
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "O"]
    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and
                   dataframe[col].dtypes != "O"]
    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and
                   dataframe[col].dtypes == "O"]
    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]

    # num_cols
    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "O"]
    num_cols = [col for col in num_cols if col not in num_but_cat]
    num_cols=  [col for col in num_cols if col not in df.select_dtypes(include=['datetime64']).columns] #ekledim

    print(f"Observations: {dataframe.shape[0]}")
    print(f"Variables: {dataframe.shape[1]}")
    print(f'cat_cols: {len(cat_cols)}')
    print(f'num_cols: {len(num_cols)}')
    print(f'cat_but_car: {len(cat_but_car)}')
    print(f'num_but_cat: {len(num_but_cat)}')
    return cat_cols, num_cols, cat_but_car

```

```python
def high_correlated_cols (dataframe, plot=False, corr_th=0.90):
    corr = dataframe.corr()
    cor_matrix = corr.abs()
    upper_triangle_matrix = cor_matrix.where(np.triu(np.ones(cor_matrix.shape), k=1).astype(np.bool_))
    #yüksek korelasyonlu değişkenler için
    drop_list = [col for col in upper_triangle_matrix.columns if any(upper_triangle_matrix[col] > corr_th)]
    if plot:
        #import seaborn as sns
        #import matplotlib
        #matplotlib.use("Qt5Agg")
        #from matplotlib import pyplot as plt
        sns.set_theme(rc={"figure.figsize": (15, 15)})
        sns.heatmap(corr, cmap="RdBu",annot=True)
        plt.show()
    return drop_list
```

```python
def outlier_thresholds(dataframe, col_name, q1=0.0, q3=0.85):
    quartile1 = dataframe[col_name].quantile(q1)
    quartile3 = dataframe[col_name].quantile(q3)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit, up_limit
```

```python
def visualize_outliers(dataframe, num_cols):
    n_cols = len(num_cols)
    fig, axes = plt.subplots(2, n_cols, figsize=(4 * n_cols, 8))
    fig.suptitle('Aykırı Değer Analizi', fontsize=14, fontweight='bold')

    if n_cols == 1:
        axes = axes.reshape(-1, 1)

    for i, col in enumerate(num_cols):
        # Aykırı değerleri tespit et
        low_limit, up_limit = outlier_thresholds(dataframe, col)
        outliers_mask = (dataframe[col] < low_limit) | (dataframe[col] > up_limit)
        clean_data = dataframe[~outliers_mask][col]

        # Scatter plot
        colors = ['red' if outlier else 'blue' for outlier in outliers_mask]
        axes[0, i].scatter(range(len(dataframe)), dataframe[col], c=colors, alpha=0.6, s=20)
        axes[0, i].axhline(up_limit, color='orange', linestyle='--', alpha=0.8)
        axes[0, i].axhline(low_limit, color='orange', linestyle='--', alpha=0.8)
        axes[0, i].set_title(f'{col}\nAykırı: {outliers_mask.sum()}')
        axes[0, i].grid(True, alpha=0.3)

        # Box plot karşılaştırması
        box_data = [dataframe[col], clean_data]
        bp = axes[1, i].boxplot(box_data, labels=['Orijinal', 'Temiz'], patch_artist=True)
        bp['boxes'][0].set_facecolor('red')
        bp['boxes'][1].set_facecolor('blue')
        axes[1, i].set_title(f'Ort: {dataframe[col].mean():.1f}→{clean_data.mean():.1f}')
        axes[1, i].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()
```

```python

# 1) Tarihleri datetime'a çevir
df["searchDate"] = pd.to_datetime(df["searchDate"], errors="coerce")
df["flightDate"] = pd.to_datetime(df["flightDate"], errors="coerce")

# 2) Geçersiz tarihleri temizle
df = df.dropna(subset=["searchDate", "flightDate"]).copy()

# 3) days_before_flight
df["days_before_flight"] = (df["flightDate"] - df["searchDate"]).dt.days

# negatif olanlar data hatası → at
df = df[df["days_before_flight"] >= 0].copy()

# 4) Route
df["route"] = df["startingAirport"].astype(str) + "-" + df["destinationAirport"].astype(str)

# 5) flight_month (YYYY-MM)
df["flight_month"] = df["flightDate"].dt.to_period("M").astype(str)

# 6) Lead time bucket
bins = [-1, 3, 7, 14, 30, 60, 10**9]
labels = ["0-3", "4-7", "8-14", "15-30", "31-60", "61+"]

df["lead_time_bucket"] = pd.cut(df["days_before_flight"], bins=bins, labels=labels)

# 7) Airline (ilk segment)
df["airline"] = df["segmentsAirlineName"].astype(str).str.split(r"\|\|").str[0]
df["airline_code"] = df["segmentsAirlineCode"].astype(str).str.split(r"\|\|").str[0]

# 8) Price numeric + basic sanity
df["totalFare"] = pd.to_numeric(df["totalFare"], errors="coerce")
df["baseFare"] = pd.to_numeric(df["baseFare"], errors="coerce")
df = df.dropna(subset=["totalFare"]).copy()
df = df[df["totalFare"] > 0].copy()

print("Feature engineering bitti. Son shape:", df.shape)
df[["legId","searchDate","flightDate","days_before_flight","lead_time_bucket","route","flight_month","airline","totalFare"]].head()
```

```python
price_df = df[
    [
        "legId",
        "searchDate",
        "flightDate",
        "days_before_flight",
        "lead_time_bucket",
        "route",
        "airline",
        "totalFare"
    ]
].copy()

global_timing = (
    price_df
    .groupby("lead_time_bucket")
    .agg(
        flight_count = ("legId", "count"),
        avg_price    = ("totalFare", "mean"),
        median_price = ("totalFare", "median"),
        min_price    = ("totalFare", "min"),
        max_price    = ("totalFare", "max"),
        std_price    = ("totalFare", "std"),
    )
    .reset_index()
)
bucket_order = ["0-3","4-7","8-14","15-30","31-60","61+"]

global_timing["lead_time_bucket"] = pd.Categorical(
    global_timing["lead_time_bucket"],
    categories=bucket_order,
    ordered=True
)

global_timing = global_timing.sort_values("lead_time_bucket")

global_timing
```

```python
global_timing = global_timing[global_timing["lead_time_bucket"] != "61+"].copy()


# 60+ lead_time kısmını temizliyoruz Nan
bucket_order = ["0-3","4-7","8-14","15-30","31-60"]

price_df = price_df[price_df["lead_time_bucket"].isin(bucket_order)].copy()
```

```python
global_timing.to_csv(output_path+"early_search.csv", index=False)
```

```python
route_timing = (
    price_df
    .groupby(["route", "lead_time_bucket"], observed=True)
    .agg(
        flight_count=("legId","count"),
        avg_price=("totalFare","mean"),
        median_price=("totalFare","median"),
        min_price=("totalFare","min"),
        max_price=("totalFare","max"),
        std_price=("totalFare","std"),
    )
    .reset_index()
)

route_timing["lead_time_bucket"] = pd.Categorical(
    route_timing["lead_time_bucket"],
    categories=bucket_order,
    ordered=True
)

route_timing = route_timing.sort_values(["route","lead_time_bucket"])
route_timing.head(30)
```

```python
# last minute price increase yani son dakika alımda bir fiyat artısı var mı buna bakıyoruz
lead_time_price_pct = route_timing.pivot(
    index="route",
    columns="lead_time_bucket",
    values="median_price"
).reset_index()

lead_time_price_pct["last_minute_price_increase_pct"] = (
    (lead_time_price_pct["0-3"] / lead_time_price_pct["8-14"] - 1) * 100
)
lead_time_price_pct["early_vs_late_pct"] = (
    (lead_time_price_pct["31-60"] / lead_time_price_pct["0-3"] - 1) * 100
)

lead_time_price_pct.sort_values("last_minute_price_increase_pct", ascending=False).head(10)
```

```python
optimal_bucket = (
    route_timing
    .dropna(subset=["median_price", "lead_time_bucket"])
    .sort_values(["route", "median_price"])
    .groupby("route", as_index=False)
    .first()
)

optimal_bucket = optimal_bucket.rename(columns={
    "lead_time_bucket": "cheapest_lead_time_bucket",
    "median_price": "cheapest_median_price"
})

optimal_bucket.head()
```
